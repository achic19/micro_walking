```python
import os
import geopandas as gpd
import momepy
from _datetime import datetime
import networkx as nx

from networkx.algorithms.centrality import edge_betweenness_centrality
```

```python
# Convert our network to a graph
general_path = os.path.dirname(os.path.abspath(os.getcwd()))
network_data = gpd.read_file(f'{general_path}/output/streets_elements/streets_elements.shp')
graph_pr = momepy.gdf_to_nx(network_data, approach="primal", length='length')
```

```python
# Calculate centrality -- closeness
print('calculate closeness at {}'.format(datetime.now()))
# line_graph convert graph to line graph so edges become nodes
edge_centrality = nx.closeness_centrality(nx.line_graph(graph_pr))
nx.set_edge_attributes(graph_pr, edge_centrality, 'closeness')
print("finish closeness- {}".format(datetime.now()))
```

```python
# Calculate centrality -- betweenness
print("calculate betweenness- {}".format(datetime.now()))
dic = edge_betweenness_centrality(graph_pr)
nx.set_edge_attributes(graph_pr, dic, 'betweenness')
```

```python
# export file with centrality measures
nodes, edges, W = momepy.nx_to_gdf(graph_pr, spatial_weights=True)
edges.to_file(f'{general_path}/output/streets_elements/streets_elements_2.shp')
```
